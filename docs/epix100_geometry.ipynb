ePix100 geometry
---------------------

ePix100 detectors have one module of 704 Ã— 768 pixels. Module built from 4 ASICs with 352 rows and 384 columns of pixels. The pixels on the inner edges of ASICs are bigger.

There are 4 more rows in raw data. These are calibration pixels. They have the same electronics as normal pixels but aren't wired to the sensor. They are two first and two last rows in the raw data array. Class `Epix100Geometry` assumes that calibration rows are cut.

```python
%matplotlib inline
from extra_geom import Epix100Geometry
```

Idealized ePix100 geometry layout may be created without any parameters. You may specify the origin. By default, the origin is at the detector center.

```python
geom = Epix100Geometry.from_origin([0, 0])
geom.inspect()
```

We can also load JUNGFRAU geometry from a [CrystFEL format geometry file](https://www.desy.de/~twhite/crystfel/manual-crystfel_geometry.html). These are text based files, often saved with a .geom extension.

```python
# Create a CrystFEL format geometry file to demonstrate loading it
geom.write_crystfel_geom('epix100_example.geom', adu_per_ev=1, clen=1)
```

```python
# Load geometry from CrystFEL format file
geom2 = Epix100Geometry.from_crystfel_geom('epix100_example.geom')
geom2.inspect()
```
